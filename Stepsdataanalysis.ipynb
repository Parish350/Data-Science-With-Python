```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("steps.csv")
x = data ['Day']
y = data ['Steps']
plt.bar(x,y)
plt.axhline(10000)
plt.xlabel("Day")
plt.ylabel("Steps")
plt.title("Daily Step Count(90 Days)")
plt.show()
```
