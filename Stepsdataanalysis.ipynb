```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("steps.csv")
x = data ['Day']
y = data ['Steps']
plt.bar(x,y)
plt.axhline(10000)
plt.xlabel("Day")
plt.ylabel("Steps")
plt.title("Daily Step Count(90 Days)")
plt.show()
```

```python
data = pd.read_csv("steps.csv")
x = data ['Day']
y = data ['Steps']
plt.bar(x,y)
plt.axhline(11000)
plt.xlabel("Day")
plt.ylabel("Steps")
plt.title("Daily Step Count (100 Days)")
plt.show()
```

```python
data["week"] = (data["Day"] - 1) //7+ 1
week_avg = data.groupby("week")["Steps"].mean()
plt.bar(week_avg.index, week_avg.values)
plt.xlabel("week")
plt.ylabel("number of steps")
plt.title(" Weekly average(100 Days)")
plt.show()
```
