```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
Y1 = [2,4,6,8,10]
plt.plot(x, Y1, "ro", linewidth = 4, label = 'y = x')
plt.plot(x, y, "ro", linewidth = 4, label = 'y = 2x')
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")
plt.title("Sample Graph")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
Y1 = [2,4,6,8,10]
Y2 = [3,6,9,12,15]
Y3 = [4,8,12,16,20]
plt.plot(x, Y1, "g^", linewidth = 4, label = 'y = x')
plt.plot(x, y, "b--", linewidth = 4, label = 'y1 = x')
plt.plot(x, Y2, "r--", linewidth = 4, label = 'y2 = x')
plt.plot(x, Y3, "b-", linewidth = 4, label = 'y3 = x')
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")
plt.title("Sample Graph")
plt.legend()
plt.show()
```
