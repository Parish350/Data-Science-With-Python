```python
data = pd.read_csv("titanic.csv")
#print(data)
print(data.head())
#Gives the typical summary of data
print(data.info())
```

```python
#filtering data based on age
above35 = data[data["Age"]>35]
print(above35.info())
#filter the age above 60 years
above60 = data[data["Age"]>60]
print(above60.info())
```

```python
maleData = data[data["Sex"] == "male"]
print(maleData.info())
femaleData = data[data["Sex"] == "female"]
print(femaleData.info())
```

```python
maleFirstClass = data[(data["Sex"] == "male") & (data["Pclass"] == 1)]
print(maleFirstClass.info())
femaleSecondClass = data[(data["Sex"] == "female") & (data["Pclass"] == 2)]
print(femaleSecondClass.info())
```

```python
female2and3 = data[(data["Sex"] == "female") & (data["Pclass"] == 2)|(data["Pclass"] == 3)]
print(female2and3.info())
```

```python
print(data["Age"].mean())
print(data["Fare"].mean())
```

```python
print(data.agg({"Age":["min","max","median"],"Fare":["min","max","median"]}))
```

```python
#Renaming the column names
data_renamed = data.rename(columns = {"Parents/Children Aboard": "New"})
print(data_renamed.columns)
```

```python
#Group by data categorically
print(data[["Pclass","Fare"]].groupby("Pclass").mean())
```
